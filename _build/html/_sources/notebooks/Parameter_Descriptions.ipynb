# Parameter descriptions

Below is a list of variables available in the SAR datasets with some adherent descriptions.

### sensor_view_angle

The angle between a sensor's line of sight and the vertical downward direction. A zero angle indicates the sensor is looking directly downward. 

### incidence_angle 

The angle that a ray coming from a light source makes with a line perpendicular to the surface at the point of incidence.  


### sensor_azimuth

The horizontal angle measured clockwise from north to the direction the sensor is pointing.


### sigmaO_VV

The scattering/absorption/attenuation coefficient is assumed to cover all wavelengths, unless specified by the the radiation_wavelength coordinate. It pertains to the radiation's deflection without energy loss. Backwards scattering involves scattering into all backward angles, i.e. those exceeding pi/2 radians. This quantity should not specify a scattering_angle. 


The radar cross section is a measure of the detectability of an object. It's the ratio of the power scattered by the object in a specific direction to the power that would be scattered by an ideal reflector.


### topographic_height 

The vertical distance or height of a point above sea level. 


### valid_land_doppler

Valid land pixels (pixels near water boundaries have been masked)


### valid_sea_doppler

Valid water covering pixels (near land boundaries have been masked) 


### valid_doppler 

Valid Doppler pixels (over land, ocean or ice).

### fe

Doppler frequency shift due to electronic mispointing.

### fgeo

Doppler frequency shift due to orbit geometry.

### fdg

Radar Doppler frequency shift due to surface velocity.

### fww

Radar Doppler frequency shift due to wind waves.

### std_fww

Standard deviation of radar Doppler frequency shift due to wind waves.

### u_range

Sea surface current velocity in range direction.

### std_u_range

Standard deviation of sea surface current velocity in range direction.

### wind_direction

SAR look relative wind from direction used in CDOP calculation.

### wind_speed

Wind speed used in CDOP calculation.

### longitude

Longitude degrees east.

### latitude

Latitude degrees north.
